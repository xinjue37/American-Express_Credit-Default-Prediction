# Data Processing

```python
import numpy as np
import pandas as pd
```

```python
df_train1013 = np.array(pd.read_csv("train_1013.csv"))
df_label1013 = np.array(pd.read_csv("label_1013.csv"))
df_train59 = np.array(pd.read_csv("train_59.csv"))
df_label59 = np.array(pd.read_csv("label_59.csv"))
df_train14 = np.array(pd.read_csv("train_14.csv"))
df_label14 = np.array(pd.read_csv("label_14.csv"))
```

```python
from sklearn.model_selection import train_test_split
xtrain1013, xtest1013, ytrain1013, ytest1013 = train_test_split(df_train1013, 
                                                                df_label1013, 
                                                                train_size=0.8,
                                                                random_state=42)
```

```python
xtrain59, xtest59, ytrain59, ytest59 = train_test_split(df_train59, 
                                                        df_label59, 
                                                        train_size=0.8,
                                                        random_state=42)
```

```python
xtrain14, xtest14, ytrain14, ytest14 = train_test_split(df_train14, df_label14, 
                                                        train_size=0.8,
                                                        random_state=42)
```

# Build the model

```python
def clf(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    pred_prob = model.predict_proba(X_test)
    
    return y_pred, pred_prob
```

```python
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix
def evaluation_score(y_pred, y_test, pred_prob, group):
    print(f"Confusion matrix for {group}: {confusion_matrix(y_pred, y_test)}")
    print(f"f1 score for {group}: {f1_score(y_pred, y_test)}")
    print(f"Accuracy score {group}: {accuracy_score(y_pred, y_test)}")
    
    print(f"Average Probability of getting '0'for {group}: {pred_prob[:, 0].mean()}")
    print(f"Average Probability of getting '1' for {group}: {pred_prob[:, 1].mean()}")
```

### i) XGBoost for 1013

```python
%%time
from xgboost import XGBClassifier
import xgboost as xgb

xgb_all_1013 = XGBClassifier()
pred1013all, predprob1013all = clf(xgb_all_1013, xtrain1013, ytrain1013, xtest1013, ytest1013)
```

```python
evaluation_score(pred1013all, ytest1013, predprob1013all, "10-13")
```

```python
xgb_all_1013.save_model("xgb_1013.json")
```

```python
XGBoost_model = xgb.XGBClassifier()
XGBoost_model.load_model('xgb_1013.json')
```

```python
XGBoost_model.predict(xtest1013)
```

```python
XGBoost_model.predict_proba(xtest1013)
```
